```python
import os
# files = os.listdir('final_regcbmJsons')
files = os.listdir('updated')
files = [f for f in files if 'json' in f and 'regcbm' in f]
print(files)
```

```python
import json
data = []
for f in files:
    if 'regcbm' in f:
        with open(f'updated/{f}', 'r') as file:
            data.append(json.load(file))
```

```python
data
```

```python
alphas = [float(f.replace('__','_').split('_')[2]) for f in files]
ps = [float(f.replace('__','_').split('_')[3].replace('.json', '')) for f in files]
print(len(alphas), len(ps), len(data))
df = pd.DataFrame({'coverage': [float(d['coverage'][:2]) for d in data], 'system accuracy': [d['system accuracy'] for d in data], 'expert accuracy': [d['expert accuracy'] for d in data], 'classifier accuracy': [d['classifier accuracy'] for d in data], 'alone classifier': [d['alone classifier'] for d in data], 'alpha': alphas, 'prob': ps})
```
